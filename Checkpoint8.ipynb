```python
import numpy as np
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from   sklearn.cluster import AgglomerativeClustering, KMeans
import scipy.cluster.hierarchy as sch
from   scipy.cluster.hierarchy import dendrogram, linkage
```

```python
df = pd.read_csv('CC GENERAL.csv')
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.loc[(df['MINIMUM_PAYMENTS'].isnull()==True),'MINIMUM_PAYMENTS'] = df['MINIMUM_PAYMENTS'].median()
df.loc[(df['CREDIT_LIMIT'].isnull()==True),'CREDIT_LIMIT'] = df['CREDIT_LIMIT'].median()
```

```python
df.isnull().sum()
df=df.drop("CUST_ID",axis=1)
```

```python
model=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(df)  
```

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(15,6))
dendrogram = sch.dendrogram(sch.linkage(df.iloc[:,[0,2]].values, method = 'ward'))
plt.title('Dendrogram', fontsize = 20)
plt.xlabel('BALANCE AND PURCHASES')
plt.ylabel('Euclidean  Distance')
plt.show()
```

```python
scaler = StandardScaler()
creditcard_df_scaled = scaler.fit_transform(df)
```

```python
kmeans = KMeans(n_clusters=6,random_state=0)
kmeans.fit(df)
```

```python
kmeans.labels_
```

```python
Sum_of_squared_distances = []
K = range(1,21)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(df)
    Sum_of_squared_distances.append(km.inertia_)
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```
